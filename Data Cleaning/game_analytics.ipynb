```python
import pandas as pd
import google.colab.files as files
import io


# 1. Đọc file b
uploaded_files = files.upload()

file_name = next(iter(uploaded_files))
df = pd.read_csv(io.BytesIO(uploaded_files[file_name]))
```

```python
df.head()
```

```python


# 2. Chuyển đổi cột event_date sang định dạng DateTime chuẩn
df['event_date'] = pd.to_datetime(df['event_date'], format='%Y%m%d')

# 3. Xử lý giá trị Null
df['revenue'] = df['revenue'].fillna(0)

# Với level, gán 0 cho những sự kiện không có level (như mở app)
df['level'] = df['level'].fillna(0).astype(int)

# 4. Gắn nhãn phân khúc người chơi dựa trên Revenue (Whale, Dolphin, Minnow)
def segment_users(rev):
    if rev > 50: return 'Whale'
    elif rev > 10: return 'Dolphin'
    elif rev > 0: return 'Minnow'
    else: return 'Non-Payer'

user_revenue = df.groupby('user_pseudo_id')['revenue'].sum().reset_index()
user_revenue['user_segment'] = user_revenue['revenue'].apply(segment_users)

# Merge nhãn phân khúc ngược lại vào bảng chính
df = df.merge(user_revenue[['user_pseudo_id', 'user_segment']], on='user_pseudo_id', how='left')

# 5. Xuất file sạch để đưa vào Power BI
df.to_csv('flood_it_clean.csv', index=False)
print("Làm sạch xong! File flood_it_cleaned.csv đã sẵn sàng.")
```
